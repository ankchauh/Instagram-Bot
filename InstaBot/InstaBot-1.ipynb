## Searching 'food' and printing all Instagram handles in the list

```python
def search(string):
    #clicking search button
    search_button=driver.find_element('xpath',
                '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div/a/div'
                       )
    search_button.click()
    #waiting for popup
    search_input=WebDriverWait(driver,5).until(ec.visibility_of(driver.find_element('xpath',
                   '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[1]/div/div/input'
                                                                      )))
    #entering string in search box
    search_input.send_keys(string)
    #wait
    time.sleep(3)
    #collecting search data
    handles=[]
    all_search=driver.find_element('xpath',
                                   '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[2]/div'
                                  )
    all_search_items=all_search.find_elements('tag name','a')
    for i in all_search_items:
        if (len(i.text)!=0) and (i.text.startswith('#')==False) and (' ' not in i.text.splitlines()[0]):
            handles.append(i.text.splitlines()[0])
    search_button.click()
    return handles
```

```python
food_handles=search('food')
for i in food_handles:   
    print(i)
```

## Like/Unlinke top 30 posts(the below functions can like/unlike any number of posts mentioned in argument)

```python
def likePosts(account, noOfPosts):
    searchAndOpen(account)
    driver.implicitly_wait(5)
    driver.execute_script(f"window.scrollTo(0, 300);")
    count=0
    driver.implicitly_wait(5)
    l=WebDriverWait(driver,5).until(ec.visibility_of(driver.find_element('css selector','div._ac7v._al3n')))
    l1=l.find_elements('css selector','div._aabd._aa8k._al3l')
    for post in l1:
        post.click()
    
        like_button=WebDriverWait(driver,5).until(ec.visibility_of(driver.find_element('css selector','span._aamw')))
        if like_button.find_element('tag name','svg').get_attribute('aria-label') == 'Like':
            like_button.click()
        else:
            print(f"You have already liked the post number-{count+1}")
        count+=1
        close_button=WebDriverWait(driver,5).until(ec.visibility_of(driver.find_element('xpath','/html/body/div[8]/div[1]/div/div[2]/div/div')))
        close_button.click()
        driver.implicitly_wait(2)
    curr_h=0
    while count<noOfPosts:
        
        try:
            l=l.find_element(By.XPATH,'following-sibling::*[1]')
            l1=l.find_elements('css selector','div._aabd._aa8k._al3l')
            for post in l1:
                post.click()

                like_button=WebDriverWait(driver,5).until(ec.visibility_of(driver.find_element('css selector','span._aamw')))
                if like_button.find_element('tag name','svg').get_attribute('aria-label') == 'Like':
                    like_button.click()
                else:
                    print(f"You have already liked the post number-{count+1}")
                count+=1
                driver.implicitly_wait(3)
                close_button=WebDriverWait(driver,5).until(ec.visibility_of(driver.find_element('xpath','/html/body/div[8]/div[1]/div/div[2]/div/div')))
                close_button.click()
                driver.implicitly_wait(2)
        except:
            a=driver.execute_script("return document.body.scrollHeight")
            driver.execute_script(f"window.scrollTo({curr_h}, {curr_h+600});")
            b=driver.execute_script("return document.body.scrollHeight")
            curr_h+=600
            if a==b:
                return
```

## Extracting list of followers

```python
def get_followers(account,number):
    if number==0:
        return
    searchAndOpen(account)
    driver.implicitly_wait(5)
    followers_button=driver.find_element('xpath',
                    '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a'
                   )
    followers_button.click()
    driver.implicitly_wait(5)
    followers=[]
    follower=driver.find_element('css selector','div.x1dm5mii.x16mil14.xiojian.x1yutycm.x1lliihq.x193iq5w.xh8yej3')
    followers.append(follower.find_element('tag name','span').text)
    count=1
    while count<number:
        try:
            follower=follower.find_element(By.XPATH,'following-sibling::*[1]')
            if follower.find_element('tag name','span').text!='':
                followers.append(follower.find_element('tag name','span').text.splitlines()[0])
                count+=1
        except NoSuchElementException:
            pop_up=driver.find_element('css selector','div._aano')
            a=driver.execute_script('return arguments[0].scrollHeight;', pop_up)
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;', pop_up)
            time.sleep(2)
            b=driver.execute_script('return arguments[0].scrollHeight;', pop_up)
            if a==b:
                break
    return followers
```

## Extract the followers of an account that you are following but they don't follow you

```python
def get_own_followers():
    profile=driver.find_element('xpath',
        '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[2]/section/main/div[1]/div[2]/div/div[1]/div/div/div/div/div'
                   )
    profile.find_element('tag name','a').click()
    driver.implicitly_wait(3)
    followers_button=driver.find_element('xpath',
                    '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a'
                   )
    followers_button.click()
    followers=[]
    driver.implicitly_wait(5)
    follower=driver.find_element('css selector','div.x1dm5mii.x16mil14.xiojian.x1yutycm.x1lliihq.x193iq5w.xh8yej3')
    followers.append(follower.find_element('tag name','span').text)
    count=1
    while True:
        try:
            follower=follower.find_element(By.XPATH,'following-sibling::*[1]')
            followers.append(follower.find_element('tag name','span').text)
            count+=1
        except NoSuchElementException:
            pop_up=driver.find_element('css selector','div._aano')
            a=driver.execute_script('return arguments[0].scrollHeight;', pop_up)
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;', pop_up)
            time.sleep(2)
            b=driver.execute_script('return arguments[0].scrollHeight;', pop_up)
            if a==b:
                break
    return followers
```

```python
def extract_target_followers(account):
    own_followers=get_own_followers()
    driver.back()
    driver.back()
    searchAndOpen(account)
    driver.implicitly_wait(5)
    followers_button=driver.find_element('xpath',
                    '/html/body/div[2]/div/div/div[2]/div/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a'
                   )
    followers_button.click()
    driver.implicitly_wait(5)
    followers=[]
    follower=driver.find_element('css selector','div.x1dm5mii.x16mil14.xiojian.x1yutycm.x1lliihq.x193iq5w.xh8yej3')
    try:
        if follower.find_element('css selector','div._ap3a._aaco._aacw._aad6._aade').text=='Following':
            followers.append(follower.find_element('tag name','span').text)
    except NoSuchElementException:
        pass
    while True:
        try:
            follower=follower.find_element(By.XPATH,'following-sibling::*[1]')
            if follower.find_element('css selector','div._ap3a._aaco._aacw._aad6._aade').text=='Following':
                followers.append(follower.find_element('tag name','span').text)
            else:
                break
        except NoSuchElementException:
            pop_up=driver.find_element('css selector','div._aano')
            a=driver.execute_script('return arguments[0].scrollHeight;', pop_up)
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight;', pop_up)
            time.sleep(2)
            b=driver.execute_script('return arguments[0].scrollHeight;', pop_up)
            if a==b:
                break
    ans=[]
    for i in followers:
        if i not in own_followers:
            ans.append(i)
    return ans
```

```python
l=extract_target_followers('foodtalkindia')
for i in l:
    print(i)
```

## Watch story and print suitable message

```python
def watch_story(account):
    searchAndOpen(account)
    driver.implicitly_wait(5)
    story_button=driver.find_element('css selector','div._aarf._aarg')
    if story_button.get_attribute('aria-disabled')=='true':
        print('Account has no story')
        return
    elif story_button.find_element('tag name','canvas').get_attribute('height')=='249':
        print('You have already seen the story')
        return
    story_button.click()
    print('You have opened the story')
```
